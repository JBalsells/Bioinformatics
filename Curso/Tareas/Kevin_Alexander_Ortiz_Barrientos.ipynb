Luego ejecutamos la funcion con la cadena data y para cada nucle√≥tido y asignamos lo que retorna a las variables A para adenina y l para longitud. Esto igual para los otros tres ncle√≥tidos. Luego imprimimos el porcentaje pero previo lo convertimos a str para poderlo imprimir. Tambi√©n se imprime la longitud y la suma de los porcentajes retornados para cada nucle√≥tido.

```python
def porcentaje_de_nucleotidos(cadena,nucleotido):
  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  porcentaje = cantidad_nucleotido*100/longitud

  return porcentaje, longitud
```

#Problema 4

En una funcion con los parametros cadena y subcadena definimos una lista llamada indices. 

Luego utilizando un bucle for indicamos que dentro del rango desde la posici√≥n 0 hasta la posici√≥n del largo de la cadena menos la subcadena. En la variable subcadena_buscada asignamos una lista que va desde i(inicia en posicion 0) hasta i+largo de subcadena. Luego le indicamos que si subcaena es la misma que subcadena buscada lo agregue a la lista indice y por √∫ltimo le pedimos quenos retorne la lista indices.

Al ejecutar la funci√≥n nos devuelve la posici√≥n donde esta la subcadena buscada.

```python
def buscar_indices(cadena, subcadena):
  indices = []
  for i in range(0,len(cadena)-len(subcadena)+1):
    subcadena_buscada = cadena[i:i+len(subcadena)]
    if(subcadena == subcadena_buscada):
        indices.append(i)
  return indices
```

luego ejecutamos la funcion primer funci√≥n y lo asignamos a la variable datos separados.

se imprime la variable datos separados y luego se imprime la segunda funci√≥n donde el primer par√°metro es la lista datos separados.

```python
def split_equal_strings(cadena, substrings):
  split_data = []
  k = len(cadena)//substrings

  for i in range (0,len(cadena),k):
    split_data.append(cadena[i:i+k])
  return split_data

def percentaje_analysis(split_array, nucleotide):
  perc_data = []
    
  for i in range(0,len(split_array)):
   long = len(split_array[i])
   coun = split_array[i].count(nucleotide)

   perc = coun*100/long
   perc_data.append(perc)
  return perc_data
```

```python
datos_separados = split_equal_strings(data,10)
print(datos_separados)
print(percentaje_analysis(datos_separados,"A"))
```

*   Obtener la cantidad de nucleÃ³tidos que tiene una cadena en total
*   Obtener el porcentaje de adeninas, guaninas, citosinas y timinas que contiene la cadena.



```python
def porcentaje_de_nucleotidos(cadena,nucleotido):

  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  porcentaje = cantidad_nucleotido*100/longitud

  return porcentaje, longitud
```

# Problema 4

**Funcion:** encontrar los Ã­ndices de una cadena en
los cuales inicia una subcadena(o palabra) almacenada en una variable.


*  **Nombre de la funciÃ³n:** buscar_indices
*   **ParÃ¡metros a ingresar:** cadena de nucleÃ³tidos, y palabra a buscar dentro de
la cadena.
*   **Retornar: **arreglo de Ã­ndices dÃ³nde se encuentra el inicio de cada palabra.






```python
def buscar_indices(cadena, subcadena): 
  indices = []
  for i in range(0,len(cadena)- len(subcadena)+1):
      subcadena_buscada = cadena [i:i+len(subcadena)]
      if(subcadena == subcadena_buscada):
        indices.append(i)
  return indices 
```

**FunciÃ³n:** generar una cantidad de subcadenas de
n nucleÃ³tidos a partir de una cadena principal y obtener el porcentaje de nucleÃ³tidos
que se encuentran en cada una de las subcadenas generadas. Luego de esto, almacene los
porcentajes de manera ordenada en una lista.

**Nombre de la funciÃ³n:** split_equal_strings.

**ParÃ¡metros a ingresar:** cadena de nucleÃ³tidos, cantidad de subcadenas a generar.

**Retornar:** lista de porcentajes de nucleÃ³tidos encontrados.

```python
def split_equal_strings(cadena, numero_de_subcadenas):
  lista_porcentajes = []
  E = len(cadena)//numero_de_subcadenas

  for i in range(0,len(cadena),E):
    lista_porcentajes.append(cadena[i:i+E])
  return lista_porcentajes
def analisis_porcentaje (split_array, nucleotido):
  perc_data = []

  for i in range (0,len(split_array)):
    long = len(split_array[i])
    count = split_array[i].count(nucleotido)

    porcentaje = count*100/long
    perc_data.append(porcentaje) 

  return perc_data 
```

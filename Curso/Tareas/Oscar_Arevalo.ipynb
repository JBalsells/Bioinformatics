4) Indices subcadena

```python
#cadena = "AGCTGCGAGCTACGTGAGCTCGAAATAAAGCTAGCTCGATA"
cadena = choleraeFast
tmpCad = ""
if len(cadena) == 0:
  tmpCad = choleraeFast
else:
  tmpCad = cadena

def buscar_indices(cad, sub):
  ind = []
  lSub = len(sub)
  cont = 0
  while (lSub + cont) <= (len(cad)):
    aux = cad[cont:(lSub + cont)]
    if sub == aux:
      ind.append(cont)
    cont += 1
  return ind


print("Indices en donde aparece:", buscar_indices(tmpCad, "ACGTACTGA"))
```

```python
cadena = ""
tmpCad = ""
if len(cadena) == 0:
  tmpCad = choleraeFast
else:
  tmpCad = cadena

nuc = {}

def search_words(cad, k):
  lCad = len(cad)
  tot = k + 1
  cont = 0
  aux = ""
  while k < tot:
    while (cont + k) <= (lCad):
      aux = cad[cont: cont + k]
      if not nuc.get(aux):
        nuc[aux] = 1
      else:
        nuc[aux] += 1
      cont += 1
    cont = 0
    k += 1
    
  fin = sorted(nuc.items(), key=lambda x:x[1], reverse=True)

  print(fin)

search_words(tmpCad, 5)

print(tmpCad)
```

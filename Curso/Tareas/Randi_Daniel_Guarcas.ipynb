```python
from collections import Counter
import re

print("Author: Randi Guarcas")

# Open file uploaded
with open("/content/Vibrio_cholerae.txt") as file:
  chain = file.read()

print(chain)
```

```python
# nucleotides = string from file uploaded
# default base is C or other
def porcentaje_nucleotidos(nucleotides, base = "C"):
  nucleotideFind = nucleotides.count(base)
  chainLen = len(nucleotides)
  percent = nucleotideFind / chainLen
  return percent * 100

bases = ["A", "G", "C", "T"]
# each base get nucleotides percent
for b in bases:
  percent = porcentaje_nucleotidos(chain, b)
  print(b , "is a %" + str(percent))
```

```python
# nucleotides = string from file uploaded
# default index is 1 or greater
def buscar_complemento(nucleotides, index = 1):
  # move 100 positions
  subchain = nucleotides[index: index + 100]
  # create nucleotides map usign maketrans
  to3prime = subchain.maketrans("AGTC", "TCAG")
  # translate to reverse original chain
  DNAc = subchain.translate(to3prime) [::-1]
  return {
      "cadena": subchain,
      "complemento": DNAc
  }

complementaryChain = buscar_complemento(chain, 1)
print(complementaryChain)
```

```python
# nucleotides = string from file uploaded
# subchain is a fragment from chain 
def buscar_indices(nucleotides, subchain):
  # find start index from subchain based chain
  # it's use re package and finditer method
  indexes = [o.start() for o in re.finditer(subchain, nucleotides)]
  return indexes

indexes = buscar_indices(chain, "TTTT")
print(indexes)
```

```python
# nucleotides = string from file uploaded
# subchain is a fragment from chain 
def total_de_palabras(nucleotides, subchain):
  # count subchain in chain  
  return nucleotides.count(subchain)

total = total_de_palabras(chain, "TTTT")
print(total)
```

```python
# nucleotides = string from file uploaded
# chainQty is a quantity groups
def split_equal_strings(nucleotides, chainQty):
  # generate groups
  splits = [nucleotides[i:i+chainQty] for i in range(0, len(nucleotides), chainQty)]
  bases = ["A", "G", "C", "T"]
  
  percents = list()
  #find nucleotides percent in subgroups
  for item in splits:
    percents.append([b+" = %"+str(porcentaje_nucleotidos(item, b)) for b in bases])
    
  return percents

total = split_equal_strings(chain, 100000) #long qty to memory out prevent 
print(total)
```

```python
# nucleotides = string from file uploaded
# qty is a quantity words to find 
def search_words(nucleotides, qty):
  # find using pattern and re package
  reg = re.compile('\S{,'+str(qty)+'}')
  groups = dict(Counter(ma.group() for ma in reg.finditer(nucleotides.strip())))

  #return new dictionaty
  return dict(sorted(groups.items(), key=lambda item:item[1]))

total = search_words(chain, 6)
print(total)
```

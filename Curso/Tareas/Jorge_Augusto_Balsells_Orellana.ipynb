```python
def buscar_indices(cadena, subcadena):
  indices = []
  for i in range(0,len(cadena)-len(subcadena)+1):
    subcadena_buscada = cadena[i:i+len(subcadena)]
    if(subcadena == subcadena_buscada):
      indices.append(i)
  return indices
```

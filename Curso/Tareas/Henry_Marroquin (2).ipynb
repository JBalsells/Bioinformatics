**Problema No.2**

Este ejercicio se resolvio realizando una regla de tres sobre la longitud de la cadena con la cantidad de veces que aparece el caracter .

```python
def porcentaje_de_nucleotidos (cadena, nucleotido):
  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  porcentaje = cantidad_nucleotido*100/longitud
  return porcentaje
```

Ac√° se realiza la prueba para obtener el 100% al sumar todos los nucleotidos

```python
A = porcentaje_de_nucleotidos(data,'A')
T = porcentaje_de_nucleotidos(data,'T')
C = porcentaje_de_nucleotidos(data,'C')
G = porcentaje_de_nucleotidos(data,'G')
print(A)
print(T)
print(C)
print(G)
print(A + T+ C+ G)
```

**Problema No.3**

Este problema se resolvi√≥ aplicando la funcion para obtener una sub cadena de otra cadena a trav√©s del indice de inicio y adicionalmente se le suma 99 a la posici√≥n final para obtener 100 caracteres .

Posteriormente se utiliza un if para crear una nueva cadena realizando su complemento.

En la sentencia return se utiliza la funci√≥n de invertir la cadena.

```python
def buscar_complemento(cadena, indice):
  cadena_indexada = cadena[indice : indice+99]
  cadena_nueva = ''
  for i in cadena_indexada:
    if i == 'A' or i == 'a': #si es A
      cadena_nueva += 'T'
    elif   i == 'T' or i == 't': #sino  , si es T
      cadena_nueva += 'A'
    elif i == 'C' or i == 'c': #sino, si es C
      cadena_nueva += 'G'
    elif i == 'G' or i == 'g': #sino si es G
      cadena_nueva += 'C'
  return { "cadena":cadena_nueva[::-1] }
```

```python
buscar_complemento("AGCTGTGCGAGCGAGTGTGCAGAGCGTGAGCGTCGAGTGAGCGCGTGA", 0)
```

**Problema No.4**

Se realiza un ciclo while mientras el indice ea menor al largo de la cadena, al encontrar la coincidencia entonces se agrega ese indice al array indices y se reanuda la busqueda desde el indice final de la subcadena+1, sino se halla (-1) entonces nos redirigimos al final del ciclo.

```python
def buscar_indices (cadena, subcadena):
  indices = []
  indice_busqueda = 0;

  while (indice_busqueda < len(cadena)):
   indice_busqueda = cadena.find(subcadena, indice_busqueda)
   if(indice_busqueda == -1):     
    indice_busqueda = len(cadena)
   else:  #sino es -1 // significa que si la hall√≥
    indices.append(indice_busqueda)
    indice_busqueda += len(subcadena)

  return indices;
```

```python
buscar_indices("AGCTGCGAGCTACGTGAGCTCGAAATAAAGCTAGCTCGATA","AGCT")
```

**Problema #7**

Se crea una funci√≥n que vaya avanzando de caracter en caracter sacando subcadenas del largo dado, en el ciclo while se crean y se van insertando las nuevas palabras en el diccionaro de palabras, si ya existe se suma 1 a la palabra ya existente, sino se inserta, finalmente se retorna el dicionario.

```python
def search_words (cadena, tamanio):  
  indices = []
  indice_busqueda = 0;
  #print(len(cadena))  
  diccionario = {}

  while (indice_busqueda < len(cadena)):    
    subcadena = (cadena[indice_busqueda: indice_busqueda+tamanio: 1]);
    #print(subcadena)
    indice_busqueda = cadena.find(subcadena, indice_busqueda)
    indice_busqueda +=1

    
    if(len(subcadena) == tamanio): #Validacion adicional para que no ejecute el codigo cuando las subcadenas son menores al tama√±o designado
      if subcadena in diccionario.keys(): #Existe la palabra que se form√≥ en el diccionario?
        diccionario[subcadena]+=1;  #si existe, entonces le voy a sumar 1
      else:
        diccionario[subcadena]=1;   #no, no existe, entonces creo la palabra en el diccionario

  return diccionario;
```

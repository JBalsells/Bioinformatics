#Problema 2
Descripci贸n: se desarrolla el c贸digo de una funci贸n de python que obtenga la cantidad de nucle贸tidos en la cadena de ADN(*pocentaje_de_nucleotidos(cadena, nucleotido)*), ingresando par谩metros (*cadena, nucleotido*), y contabilizar y retornar(*return*) el porcentaje de Adeninas, Citocinas, Timinas y Guaninas de dicha cadena(*porcentaje*).

```python
def porcentaje_de_nucleotidos(cadena, nucleotido):
  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  
  porcentaje = (cantidad_nucleotido * 100) / longitud

  return porcentaje, longitud
```

```python
A, l = porcentaje_de_nucleotidos(data, "A")
C, l = porcentaje_de_nucleotidos(data, "C")
G, l = porcentaje_de_nucleotidos(data, "G")
T, l = porcentaje_de_nucleotidos(data, "T")

print("Cantidad de nucle贸tidos:", l, "nucle贸tidos.")

print("Adeninas:" + str(round(A,2)),"%")
print("Citocinas:" + str(round(C,2)),"%")
print("Guaninas:" + str(round(G,2)),"%")
print("Timinas:" + str(round(T,2)),"%")

print("Total:", A + C + G + T, "%")
```

#Problema 4
Descripci贸n: se desarrolla el c贸digo de una funci贸n(*buscar_indices*) de Python, para encontrar los 铆ndices de una subcadena o palabra de la secuencia original de ADN, utilizando ciclos (*for*), funci贸n (*range*) para localizar los 铆ndices; condicional(*if*) para comparar la subcadena o palabra, y luego returnarlos (*indices*) en un diccionario.

```python
def buscar_indices(cadena, subcadena):
  indices = []
  for i in range(0,len(cadena)-len(subcadena)+1):
    subcadena_buscada = cadena[i:i+len(subcadena)]
    if(subcadena == subcadena_buscada):
      indices.append(i)
  return indices
```

```python
buscar_indices(data, "ACGTACTGA")
```

y por ultimo se manda a imprimir print("Adeninas: " str(A))


```python
def porcentaje_de_nucleotidos(cadena,nucleotido):

  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  porcentaje = cantidad_nucleotido*100/longitud
  
  return porcentaje, longitud
```

```python
A,l = porcentaje_de_nucleotidos(data,"A")
C,l = porcentaje_de_nucleotidos(data,"C")
T,l = porcentaje_de_nucleotidos(data,"T")
G,l = porcentaje_de_nucleotidos(data,"G")

print("Adeninas: " + str(A))
print("Citosinas: " + str(C))
print("Timinas: " +  str(T))
print("Guaninas: " + str(G))

print("Longitud de cadena: " + str(l))
```

Se define una función para que busque los indices con dos parametros (cadena y subcadena) y retornaremos el indice que no es mas que una lista de caracteres. Y el for lo que hace es recorrer desde 0 hasta la longitud de la cadena y con la variable subcadena_buscada que nos define un espacio(pedazo) de la cadena a la cual vamos a comparar con la otra cadena. (subcadena_buscada = cadena[i+i+len(subcadena]) luego hacemos una comparación con un if donde se compara la subcadena y se verifica si es igual a la subcadena_buscada. y con el indices.appende agregamos un valor al final de la lista. Retornamos el valor de indice y por ultimo buscamos los indices con buscar_indices(data,"AGCT")

```python
def buscar_indices(cadena,subcadena):
    indices = []
    for i in range(0,len(cadena)-len(subcadena)+1):
      subcadena_buscada = cadena[i:i+len(subcadena)]
      if(subcadena == subcadena_buscada):
         indices.append(i)                                  
    return indices
```

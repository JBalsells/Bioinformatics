# **PROBLEMA 4**

Desarrolle una funci√≥n de python que encuentre los √≠ndices de una cadena en
los cuales inicia una subcadena(o palabra) almacenada en una variable.

a) Nombre de la funci√≥n: buscar_indices

b) Par√°metros a ingresar: cadena de nucle√≥tidos, y palabra a buscar dentro de
la cadena.

c) Retornar: arreglo de √≠ndices d√≥nde se encuentra el inicio de cada palabra.

4.jpg

Funci√≥n

```python
def buscar_indices(cadena,sub_cadena):
  indices = []
  for i in range(0,len(cadena)-len(sub_cadena)+1):
      subcadena_buscada = cadena[i:i+len(sub_cadena)]
      if(sub_cadena == subcadena_buscada):
        indices.append(i)
  return indices
```

a) Nombre de la funci√≥n: total_de_palabras

b) Par√°metros a ingresar: cadena de nucle√≥tidos, y palabra a buscar Dentro de la cadena.

c) Retornar: n√∫mero entero del total de veces que aparece la palabra en la cadena.

Ejemplo:

Funci√≥n total de palabras

```python
def total_de_palabras(cadena_de_nucleotidos ,  palabra_a_buscar):

  cantidad_palabras = []

  for i in range(0,len(cadena_de_nucleotidos)-len(palabra_a_buscar)+1):
      subcadena_buscada = cadena_de_nucleotidos[i:i+len(palabra_a_buscar)]
      if(palabra_a_buscar == subcadena_buscada):
        cantidad_palabras.append(i)
        total_subcadenas = int(len(cantidad_palabras))

  return total_subcadenas

```

# **PROBLEMA 6**

Desarrolle una funci√≥n en python que genere una cantidad de subcadenas de n nucle√≥tidos a partir de una cadena principal. Obtenga el porcentaje de nucle√≥tidos
que se encuentran en cada una de las subcadenas generadas. Almacene los porcentajes de manera ordenada en una lista.

a) Nombre de la funci√≥n: split_equal_strings.

b) Par√°metros a ingresar: cadena de nucle√≥tidos, cantidad de subcadenas a generar.

c) Retornar: lista de porcentajes de nucle√≥tidos encontrados.

```python
def split_equal_strings(cadena, numero_subcadenas ):
    sub_cadenas = []

    k = len(cadena)//numero_subcadenas  ##division con resultado entero. K es el n√∫mero de nucle√≥tidos que tendr√° cada sub_cadena

    for i in range(0,len(cadena),k):   #desde cero, hasta el valor de la longitud de la cadena, y con saltos del tama√±o de k
      sub_cadenas.append(cadena[i:i+k])

    return sub_cadenas

def porcentaje_analisis(mini_arreglo, nucleotido):
    porcentaje_datos = []
    
    for i in range(0,len(mini_arreglo)):
        long = len(mini_arreglo[i])
        count = mini_arreglo[i].count(nucleotido)

        porcentaje = count * 100 / long 
        porcentaje_datos.append(porcentaje)
    return porcentaje_datos
```

```python
print("Ingresa la cantidad de sub cadenas en las que deseas dividir la cadena de ADN ")
b = int(input())
lista_subcadenas = split_equal_strings(data,b)
print("\n ¬øCu√°l es el nucle√≥tido del que deseas saber su porcentaje en cada sub cadena? ")
c = input()

##print(lista_subcadenas[2])


if(c not in lista_subcadenas):

    print("\n" , porcentaje_analisis(lista_subcadenas,c))
else:
    print("El valor ingresado no esta en la cadena")
   

##len(lista_subcadenas[0])
```

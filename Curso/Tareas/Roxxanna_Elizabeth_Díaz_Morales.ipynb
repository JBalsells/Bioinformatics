-Se indica la operaci√≥n requerida para realizar el conteo de la cantidad de nucle√≥tidos presentes en la data con la funci√≥n count y el la operaci√≥n necesaria para calcular la cantidad de nucle√≥tidos. Se retorna el porcentaje y la longitud de la cadena. 
-Se realiza el calculo y se imprime el resultado de los 4 nucleotidos presentes en la cadena y se imprime tambi√©n el resultado de la suma de estos. 


```python
def porcentaje_de_nucleotidos(cadena, nucleotido):
  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  porcentaje = cantidad_nucleotido*100/longitud

  return porcentaje, longitud
```

```python
def buscar_indices(cadena, subcadena):
  indices = []
  for i in range(0,len(cadena)-len(subcadena)+1):
    subcadena_buscada = cadena[i:i+len(subcadena)]
    if(subcadena == subcadena_buscada):
      indices.append(i)
  return indices
```

-Se realizan dos funciones, una funci√≥n split para realizar peque√±as porciones de la cadena o subcadenas y una funci√≥n para obtener el porcentaje de nucleotidos. 

La variable N nos devuelve el n√∫mero de subcadenas como n√∫mero entero.
se realiza un ciclo que lea la data para entrar cuantos nucleotidos hay en cada subcadena. Utilizando append agregamos la cadena al final de nuestra lista.
Utilizando un ciclo for vemos la cantidad de nucleotidos que tenemos en la longitud de cada cadena y por √∫ltimo se saca el porcentaje de cada una. 



```python
def split_equal_strings(cadena,subcadena):
  split_data = []
  N = len(cadena)//subcadena

  for i in range(0,len(cadena),N):
    split_data.append(cadena[i:i+N])
  return split_data

def porcentaje_analysis(split_array, nucleotido):
  porc_data = []

  for i in range(0,len(split_array)):
    long = len(split_array[i])
    count = split_array[i].count(nucleotido)

    porc = count*100/long
    porc_data.append(porc)
  return porc_data
```

```python
datos_separados = split_equal_strings(data,10)
print(datos_separados)
print(porcentaje_analysis(datos_separados,"A"))
print(porcentaje_analysis(datos_separados,"T"))
print(porcentaje_analysis(datos_separados,"C"))
print(porcentaje_analysis(datos_separados,"G"))
```

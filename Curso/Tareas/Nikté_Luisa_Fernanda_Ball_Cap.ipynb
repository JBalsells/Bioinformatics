Obtenemos la cantidad  de datos con las que estamos trabajando para posteriormente sacar los porcentajes, esto por medio de la varible longitud en la que obtendremos la longitud de cadena.
Crearemos la variable cantidad_nucleotido en el cual obtendremos la cantidad del nucleotido de nuestro interés en la cadena, por medio de .count.
Crearemos la variable porcentaje en la cual buscamos el porcentaje de cada nucleotido, por medio de una regla de 3, en la cual multiplicaremos la cantidad de nucleotidos *100 y lo dividiremos por la longitud. Y buscaremos el porcentaje de cada uno de los nucleótidos (A,C,T y G).

```python
def porcentaje_de_nucleotidos(cadena,nucleotido):
  longitud= len(cadena)
  cantidad_nucleotido= cadena.count(nucleotido)
  porcentaje= cantidad_nucleotido*100/longitud

  return porcentaje, longitud
```

Creamos la variable nucl con valor de la cadena en la posición cambiante(i) hasta la posición de i más la longitud de la palabras.

Se crea una condición en la que si el valor de palabra es igual a el valor de nucl. Se agrega una posición a la lista de indices, esto lo logramos por medio de .append

Siendo la palabra buscada "TATATAT" de la cadena con datos contenidos en data.


```python
def buscar_indices (cadena, palabras):
  indices= []
  for i in range(0,len(cadena)-len(palabras)+1):
    nucl= cadena[i:i+len(palabras)]
    if(palabras==nucl):
      indices.append(i)
  return indices
```

Creamos una función con el nombre split_equal_strings con parámetros cadena de nucleótidos (cadena) y cantidad de subcadenas a generar (numero_subcadenas), que en este caso queremos crear 4 subcadenas.
Con esta función solo haremos la división de subcadenas.

Y por medio de .append agregaremos datos a la variable lista_porcentaje,los datos que se agregarán será la cadena en la posición i (que va ser cambiante) hasta la posición i + el valor de la variable x.}


Crearemos una segunda función con la cual obtendremos los porcentajes de cada nucleotido en cada una de las subcadenas.
Esta función tendrá el nombre de percentaje_a con parámetros arreglo (siendo los datos_separados) y nucleotide (el nucleótido que elijamos)
Creamos la variable perc_data en la cual iremos guardando los datos de los porcentajes en una lista.

Creamos un ciclo for el cual recorrera en un rango desde la posición 0 hasta la longitud del valor de datos_separados.
Determinamos una variable perc_data cuyos valores serán los porcentajes puestos en una lista.
Haremos una regla de 3 para poder obtener el porcentaje, con lo cual primero obtendremos el valor de la longitud y la cantidad de veces que tenemos el nucleótido elejido en datos_separados. 
Por medio de .append agregaremos a la lista de perc_data los porcentajes obtenidos.

Imprimimos los porcentajes de cada uno de los nucleótidos en cada arreglo.


```python
def split_equal_strings(cadena, numero_subcadenas):
  lista_porcentaje = []
  x= len(cadena)//numero_subcadenas

  for i in range(0,len(cadena),x):
    lista_porcentaje.append(cadena[i:i+x])
  return lista_porcentaje

def porcentaje_a(arreglo,nucleotide):
  perc_data =[]

  for i in range(0,len(arreglo)):
    long =len(arreglo[i])
    coun =arreglo[i].count(nucleotide)

    perc =coun*100/long
    perc_data.append(perc)
  return perc_data
  
```

```python
datos_separados = split_equal_strings(data,4)

#print(datos_separados)
print( "% Adeninas" + " " + str(porcentaje_a(datos_separados,'A')))
print( "% Guaninas" + " " + str(porcentaje_a(datos_separados,'G')))
print( "% Timinas" + " " + str(porcentaje_a(datos_separados,'T')))
print( "% Citosinas" + " " +  str(porcentaje_a(datos_separados,'C')))

```

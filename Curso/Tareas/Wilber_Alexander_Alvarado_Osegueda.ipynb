# PROBLEMA 2
Desarrolle una funcion de python para obtener la cantidad de nucleotidos que tiene una cadena en total, y obtenga el porcentaje de adeninas, guaninas y itosinas y timinas que continene la cadena

1.   nombre de funcion: porcentaje_de_nucleotidos
2.   parametros a ingresar: cadena de nucleotidos y nuleotido a buscar en el porcentaje
3. retornar: porcentaje de nucleotido calculado



```python
def porcentaje_de_nucleotidos(cadena, nucleotido):
  
  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  porcentaje = (cantidad_nucleotido/longitud)*100
  return porcentaje, longitud
  #retoma la longitud de la cadena para contar los nucleotidos y dar como resultado el porcentaje de nucleotidos
  
```

```python
A,l = porcentaje_de_nucleotidos(seq, "A")
C,l = porcentaje_de_nucleotidos(seq, "C")
G,l = porcentaje_de_nucleotidos(seq, "G")
T,l = porcentaje_de_nucleotidos(seq, "T")

print("Porcentaje Adenina: " + str(A))
print("Porcentaje Citocina: " + str(C))
print("Porcentaje Guanina: " + str(G))
print("Porcentaje Timina: " + str(T))
print("Longitud total de la cadena: " + str(l))
```

# PROBLEMA 3
Desarrolle una funcion de python para obtener solamente 100 nucleotidos de una cadena en la cual se agregue el indice de inicio y automaticamente debe obtener los 100 nucleotidos a partir de alli. Obtenga la cadena complementaria correspondiente, considerando que la lectura de cadena es de 5' a 3'.


1.   Nombre de la funcion: buscar_complemento
2.   Parametros a ingresar: cadena de nucleotidos e indice de inicio
3. retorno diccionario cadena con 100 nucleotidos y su complemento
4. Formato de diccionario: retorno = {'cadena':'complemento'}



```python
def buscar_complemento(cadena, indice):
  subcadena = (cadena[indice:indice+100]) #se crea una sub cadena desde el indice introducido + 100 posiciones
  inicio = len(subcadena)-1 #marca la longitud de la cadena -1 es decir de atras hacia adelante
  reversed = '' #variable vacia donde se agregaran los valores
  while inicio>=0: #ciclo while donde la variable inicio sea igual a cero parara el ciclo
    if(subcadena[inicio]=="A"):
      reversed += "T" #si el inicio marcado en la subvadena es igual a A se transforma en T, C a G, etc
    elif(subcadena[inicio] =="C"):
      reversed += "G"
    elif(subcadena[inicio] =="G"):
      reversed += "C"
    elif(subcadena[inicio] =="T"):
      reversed += "A"
    inicio -= 1 
  diccionario = {subcadena:reversed} #se crea un diccionario y se retorna el diccionario
  return diccionario
```

```python
prueba = buscar_complemento(seq, 25)
print(prueba)
```

# PROBLEMA 4
Desarrolle una funcion de python que encuentre los indices de una cadena en los cuales inicia una subcadena o palabra almacenada en una variable
1. nombre de la funcion: buscar_indices
2. Parametros a ingresar: Cadena de nucleotidos y palabra a buscar dentro de la cadena
3. retornar: arreglo de indices donde se encuentra el inicio de cada palabra

```python
def buscar_indices(cadena, palabra):
   indices = [] #se crea una lista
   for i in range(0, len(cadena)-len(palabra)+1): #para la variable i dentro de la longitud total menos la palabra a buscar +1 es decir mas de una vez
     buscar = cadena[i:i+len(palabra)] #se busca dentro de la cadena la longitud de la palabra es decir donde esta
     if(palabra == buscar): #si encuentra la palabra en la variable buscar lo agrega a la lista
       indices.append(i)
 
   return indices #retorna el indice donde esta la palabra
```

```python
prueba1 = buscar_indices(seq,"ACGTATT")
print(prueba1)
```

2. Parametros: cadena de nucleotidos y palabra a buscar

# PROBLEMA 6
Desarrolle una funcion en python que genera una cantidad de subcadenas de n nucleotidos a partir de una cadena principal. Obtenga el porcentaje de nucleotidos que se encuentran en cada una de las subcadenas generadas. Almacene los porcentajes de manera ordenada en una lista.
1. nombre de funcion: split_equal_strings
2. Parametros: cadena, cantidad de subcadenas
3. retornar: lista de porcentajes de nucleotidos encontrados

```python
def split_equal_strings(cadena, partes, nucleotido):
  lista = []
  #se crea una lista 
  l = len(cadena)//partes #la longitud de la cadena divida en las partes que queremos 
  for i in range(0,len(cadena), l): #para la variable i en un rango desde cero a la longitud de la cadena hasta la parte que queremos
    lista.append(cadena[i:i+l]) #se agrega a la lista la parte de la cadena hasta la misma longitud de la partes divididas
  return analisis_perc(lista, nucleotido) #retorna a una funcion
  
def analisis_perc(lista, nucleotido): 
  porcentaje_total= [] #se crea una lista 
  for k in range(0,len(lista)): #para la variable k en el rango de cero a la longitud de la lista
    long = len(lista[k]) #la variable long, la longitud de la variable k dentro de la lista
    porcentaje = lista[k].count(nucleotido) #porcentaje de la lista k para contar el nucleotido
    porcentaje_parcial= (porcentaje/long) * 100 
    porcentaje_total.append(porcentaje_parcial) #agregar a la lista el porcentaje parcial del nucleotido
  
  return porcentaje_total #retorno de la lista 
```

```python
split_equal_strings(seq, 10, "T")
```

#PROBLEMA 7
Busque la cantidad de veces que se repiten palabras de n nucleotidos dentro de la cadena vibrio_cholerae.text
1. Nombre de la funcion: search_words
2. Parametros a ingresar: cadena de nucleotidos, tama;o de palabras a buscar
3. retornar: diccionario ordenado on las palabras encontradas asociadas a la cantidad de veces que se encontraron

```python
def search_words(cadena, longitud):
  diccionario = {} #se crea un diccionario
  for i in range(0, len(cadena)-longitud+1): #para la variable i en el rango, define desde 0 a la longitud de la cadena menos la longitud de la palabra, varias veces (+1)
     buscar = cadena[i:i+longitud] #busca dentro de la cadena la variable i +la laongitud de la palabra
     condicion = buscar in diccionario #busca la condicion en el diccionario
     if condicion == False: #para la variable si es falsa seria igual a 1, es decir si la palabra y la longitud no esta, se toma como 1 vez
       diccionario[buscar] = 1
     elif condicion == True:  #si la vuelve a contar la toma como +1, y sigue sumando
       diccionario[buscar]+=1
  diccionario_final = sorted(diccionario.items(), key=lambda item:item[1]) #ordena el diccionario de forma ascendente por la cantidad de veces que aparece
  
  return diccionario_final
```

```python
search_words(seq, 5)
```

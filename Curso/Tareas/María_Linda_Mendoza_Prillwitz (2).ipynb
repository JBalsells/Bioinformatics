PROBLEMA No.2
En este problema se definió la función porcentaje de nucleotidos, con las entradas de la cadena (sec) y el nucleotido de interés (n) se calculó el porcentaje de ese nucleotido en esa cadena, en esta función se regresa una varibale P, la cual es el porcentaje de ese nucleotido.

```python
#def es para declarar una funcion
def porcentaje_de_nucleotidos(sec,n):
  total = len(sec)
  nuc = sec.count(n)
  P = nuc/total*100
  return (P)

var = "C"
var1 = "A"
var2 = "T"
var3 = "G"
resultado =porcentaje_de_nucleotidos(seq , var)
resultado1 =porcentaje_de_nucleotidos(seq , var1)
resultado2 =porcentaje_de_nucleotidos(seq , var2)
resultado3 =porcentaje_de_nucleotidos(seq , var3)
print("El porcentaje de", var, "es", resultado,"%")
print("El porcentaje de", var1, "es", resultado1,"%")
print("El porcentaje de", var2, "es", resultado2,"%")
print("El porcentaje de", var3, "es", resultado3,"%")
print(resultado + resultado1 + resultado2 + resultado3)
```

PROBLEMA NO.3
En este problema se definió una función buscar complemento, en la cual se ingresa una cadena y un índice, la cadena será del tamaño del número ingresado como índice más 100. se crearon dos variables que se irán llenando, según si se cumple las condiciones, una ira guardando la cadena de 100 nucleotidos para mostrar y la otra la cadena complementaria. la función regresará un diccionario con la cadena y la cadena complementaria.

```python
def buscar_complemento(cadena, indice):
  reversed = ""
  chain = ""
  ind = indice+100
  while ind >= indice:
    if (cadena[ind]== "A"):
      reversed += "T"
    elif (cadena[ind]== "C"):
      reversed += "G"
    elif (cadena[ind]== "G"):
      reversed += "C"
    elif (cadena[ind]== "T"):
      reversed += "A"
    chain += cadena[ind]
    ind -=1
    

  diccionario_cadena = {chain[::-1]: reversed}
  return diccionario_cadena
buscar_complemento(seq, 50)
    
```

En este problema se definió una función de buscar índice, en el cual se hace un ingreso de una cadena y una subcadena, en este se utiliza un ciclo for donde busca desde el índice 0 hasta la longitud de la cadena menos la longitud de la cadena más uno, para que pueda pasarse a la siguiente posición hasta el fin de la cadena. Se creó una variable subcadena buscada y es comparada con la subcadena, si estas son iguales aumentará con la función append el indice, que fue creado desde el inicio y regreserá la lista de los indices donde estos son encontrados.

```python
def buscar_indices(cadena, subcadena): 
  indices = []
  for i in range (0 , len(cadena) -len(subcadena)+1):
    subcadena_buscada = cadena[i:i+ len(subcadena)]
    if (subcadena == subcadena_buscada):
      indices.append(i)
  return indices


buscar_indices(seq , "AGATCCAT")

    
```

```python
def total_de_palabras(cadena, subcadena): 
  cant = 0
  for i in range (0 , len(cadena) -len(subcadena)+1):
    subcadena_buscada = cadena[i:i+ len(subcadena)]
    if (subcadena == subcadena_buscada):
      cant+=1
  return cant


total_de_palabras(seq , "AGATCCAT")
```

PROBLEMA NO.7
En este problema se definió la función search words, en la cual tendrá entrada de parametros de cadena y p, que es le tamaño de palabra a buscar. se creó un diccionario abierto para ir siendo llenado con el siguiente ciclo for. en el ciclo for este se extiende en l¿rangos desde la longitud de la cadena menos el tamaño de la palabra a buscar más uno para trasladarse en la siguiente posición. la palabra cambiara según la posición de i en el ciclo for, después se realiza una condición se guardará la palabra como llave si es nueva en el diccionario y si esta no está, se sumara 1 a esa llave. En esta se regresara el diccionario final.

```python
def search_words(cadena, p):
  dic = {}
  for i in range(len(cadena)-p+1):
    palabra = cadena[i:i+p]
    cond = palabra in dic
    if cond:
      dic[palabra]+=1
    else:
      dic[palabra]=1
    
  return(dic)

search_words(seq,2)
```

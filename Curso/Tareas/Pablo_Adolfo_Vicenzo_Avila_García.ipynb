# Problema 2
Con nuestra funci√≥n hallamos el porcentaje de cada nucleotido en nuestra cadena dada y cuantos datos hay en la cadena.

```python
def porcentaje_de_nucleotidos(cadena, nucleotido):
  longitud = len(cadena)
  cantidad_nucleotidos = cadena.count(nucleotido)
  porcentaje= cantidad_nucleotidos*100/longitud
  return porcentaje, longitud
```

```python
A,l = porcentaje_de_nucleotidos(data, 'A')
C,l = porcentaje_de_nucleotidos(data, 'C')
T,l = porcentaje_de_nucleotidos(data, 'T')
G,l = porcentaje_de_nucleotidos(data, 'G')

print("Adeninas:", A)
print("Citosinas:", C)
print("Timinas:",T)
print("Guaninas:",G)
print("Longitud de cadena:", l)
#print(A+C+T+G)
```

# Problema 3
Cambiamos el dato de a en nuestro print para indicar donde queremos que empiece la cadena, luego encontramos el complemento haciendo las conversiones especificadas en los if's seg√∫n la ley de los nucleotidos.

```python
a=9
print(data[a:a+100])
```

```python
def buscar_complemento(data, a):
  cadena = data[a:a+100]
  index = len(cadena)-1

  complemento = ""
  while index >= 0:
    if(cadena[index]=="A" or cadena[index]== "a"):
      complemento = complemento + "T"
    elif(cadena[index]=="T" or cadena[index]=="t"):
      complemento = complemento + "A"
    elif(cadena[index]=="C" or cadena[index]=="c"):
      complemento = complemento + "G"
    elif(cadena[index]=="G" or cadena[index]=="g"):
      complemento = complemento + "C"
    index = index - 1
  return(cadena, complemento)
```

```python
print(buscar_complemento(data, 15))
```

# Problema 4
Buscamos un bloque de datos concreto en la estructura de nucleotidos dadas.
O en el caso de querer usarse en otro tipo de documento se busca una palabra en especif√≠co y muestra la posici√≥n inicial de ello.

```python
def buscar_indices(cadena, subcadena):
  indices=[]
  if(subcadena == int):
    print("Dato ingresado invalido")
    return(indices)
  for i in range(0,len(cadena)-len(subcadena)+1):
    subcadena_buscada = cadena[i:i+len(subcadena)]
    if(subcadena == subcadena_buscada):
      indices.append(i)
  return(indices)
```

```python
print(buscar_indices(data, "AACG"))
#buscar_indices(data,"AAAT")
```

# Problema 6

```python
def split_equal_strings(cadena, cant_subcadenas):
  split_data=[]
  k = len(cadena)//cant_subcadenas

  for i in range(0, len(cadena),k):
    split_data.append(cadena[i: i+k])
  return(split_data)

def porcentaje_split_equal(piece_list,nucleotido):
  porcentaje_datos=[]

  for i in range(0, len(piece_list)):
    long = len(piece_list[i])
    count = piece_list[i].count(nucleotido)
    
    perc= count*100/long
    porcentaje_datos. append(perc)
  return(porcentaje_datos)
```

```python
datos_separados = split_equal_strings(data, 10)
#len(datos_separados)
print(len(datos_separados))
print(datos_separados)
#print(len(data))
print(porcentaje_split_equal(datos_separados, "C"))
```

**3) Obtener 100 nucleótidos y obtener cadena complementaria**
1. Se corta la cadena de 0 a 100
2. Se inicia a leer desde el final de la cadena hasta 0
3. Se compara A,C,G,T y se va agregando el complemento a la variable complemento para 3' a 5'
4. Se retorna el diccionario asociando la cadena y complemento a su respectivo valor

```python
def buscar_complemento(chain, start_index):
  read = chain[start_index:start_index+100]
  complemento = ""
  for i in reversed(range(len(read))):
    if read[i].upper() == 'A': complemento += 'T'
    elif read[i].upper() == 'C': complemento += 'G'
    elif read[i].upper() == 'G': complemento += 'C'
    elif read[i].upper() == 'T': complemento += 'A'
    
  return dict({"cadena":read, "complemento":complemento})

buscar_complemento("AGCTGTGCGAGCGAGTGTGCAGAGCGTGAGCGTCGAGTGAGCGCGTGA",0)
```

**4) Encontrar indices de una subcadena**
1. Se define un array auxiliar
2. Se ejecuta desde cero hasta el tamaño de la cadena, menos el tamaño de la subcadena
3. Se agrega al array la posición en la que se encontró la subcadena


```python
def buscar_indices(chain, read):
  index = []
  for i in range(0, len(chain)-len(read)+1):
    if(chain[i:i+len(read)] == read):
      index.append(i);
  return index
buscar_indices("AGCTGCGAGCTACGTGAGCTCGAAATAAAGCTAGCTCGATA", "AGCT")
```

**5) Encontrar total de subcadenas en una cadena**
1. Se inicializa el contador
2. Se ejecuta desde cero hasta el tamaño de la cadena, menos el tamaño de la subcadena
3. Se compara si la cadena del ciclo es igual a la cadena dada y se suma el contador

```python
def total_de_palabras(chain, read):
  count = 0
  for i in range(0, len(chain)-len(read)+1):
    if(chain[i:i+len(read)] == read):
      count += 1
  return count
total_de_palabras("AGCTGCGAGCTACGTGAGCTCGAAATAAAGCTAGCTCGATA", "AGCT")
```

**6) Generar subcadenas de n nucleótidos y obtener el porcentaje de nucleótidos en cada subcadena.**

1. Se obtiene el tamaño *n* sin decimales, de las subcadenas a generar
2. Se inicializa el array de porcentajes
3. Desde 0 hasta el tamño de la cadena, a cada *n* pasos
4. Se obtiene la subcadena de tamaño *n*
5. Se obtiene el porcentaje de nucleótidos en la subcadena y se agrega al array

```python
def split_equal_string(chain, reads_qty, nucleotide):
  read_size = len(chain)//reads_qty
  percentages = []
  for i in range(0, len(chain), read_size):
    subchain = chain[i:i+read_size]
    percentages.append(porcentaje_de_nucleotidos(subchain, nucleotide))
  return percentages

def porcentaje_de_nucleotidos(chain, search_for):
  total = len(chain)
  nucleotide_qty = chain.count(search_for)
  return (nucleotide_qty/total)*100

split_equal_string("AACAEFAAA", 2,'A')
```

2. En la función search_words se inicializa el diccionario de encontrados
3. Se ejecuta desde cero hasta el tamaño de la cadena, menos el tamaño de la subcadena
4. Se genera la subcadena y se verifica si ya está agregada al diccionario
5. Si ya existe se aumenta el contador de existencia, si no se agrega la existencia actual
6. Se ordena el diccionario con el número de existencias encontradas

```python
def search_words(chain, read_size):
  found = {}
  for i in range(0, len(chain)-(read_size+1)):
    subchain = chain[i:i+read_size]
    if (subchain in found) == True: found[subchain] += 1
    else: found[subchain] = 1
  return found

words = search_words(nucleotides_chain, 4)
sorted(words.items(), key=lambda item:item[1])
```

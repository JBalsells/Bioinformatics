A continuación en guardamos en 2 variables ATCG y longitud, los valores data(donde guardamos los nucleotidos), establecemos que nucleotido buscaremos en la cadena y contabilizamos cada uno ACGT para que el dato sea valido todos los valores deben sumar el 100%. Para poder obtener el porcentaje y longitud de cada uno de los nucleotidos imprimimos la longitud de la cadena y el porcentaje de cada uno de los nucleotidos. 

```python
def porcentaje_de_nucleotidos(cadena, nucleotido):

  longitud = len(cadena)
  cantidad_nucleotido = cadena.count(nucleotido)
  porcentaje = cantidad_nucleotido*100/longitud

  return porcentaje, longitud
```

Internamente en el ciclo for hacemos una comparación entre la variable subcadena buscada sea igual a la cadena de la posición o variable i hasta longitud de la subcadena (cadena[i:i+len(subcadena)]. A continuación hacer una comparación con un doble igual (==) con un condicional if y comparamos si la subcadena es igual a la subcadena buscada, esto significa que esa lista que se llama indice empieza la subcadena que buscamos en la cadena original, entonces en esa lista agregamos un valor (indices.append) que agrega un valor al final de la lista indice. Y al finalizar retornamos indice y nos devulve todas la poscione en donde encuenta ACTGC.


```python
def buscar_indices(cadena, subcadena):
  indices = []
  for i in range(0,len(cadena)-len(subcadena)+1):
    subcadena_buscada = cadena[i:i+len(subcadena)]
    if(subcadena == subcadena_buscada):
      indices.append(i)
  return indices
```

Para sacar el porcentaje del analisis en general, creamos un parametro de todos los datos ya separados(split_array) y agregamos un nucleotido al cual vamos a calcular los porcentajes, de esta manera retornamos el porcentaje de los datos. Luego creamos un variable (perc_data) o arreglo donde vamos ir guardando los porcenajes. Posteriormente buscamos el porcentaje de subcadena comenzamos corriendo un ciclo for que recorra la lista de la función anterior desde la posición 0 hasta la longitud lista (for i in range(0,len(split_array)) que estamos realizando, dentro del for realizamos establecemos una variable long (len(split_array[i]) y coun (split array[i].count(nucleotido). Luego realizamos una regla de 3 para sacar el porcentaje (coun*100/longitud) luego lo guardamos en el arreglo perc_data con un append, de esta manera guardamos su valor. Al retornar datos_separados imprimimos perc_data, luego los parametros (porcentaje_analisis), enviamos la función datos separados y el núcleotido que buscamos.

```python
def split_equal_strings(cadena, numero_subcadenas,nucleotide):
  lista_porcentajes = []
  a = len(cadena)//numero_subcadenas

  for i in range(0,len(cadena),a):
    lista_porcentajes.append(cadena[i:i+a])
  return porcentaje_analisis(lista_porcentajes,nucleotide)

def porcentaje_analisis(split_array, nucleotide):
  perc_data = []

  for i in range(0,len(split_array)):
    long = len(split_array[i])
    coun = split_array[i].count(nucleotide)

    perc = coun*100/long
    perc_data.append(perc)
  return perc_data
```
